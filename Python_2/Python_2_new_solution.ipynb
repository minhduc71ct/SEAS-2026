# Exercise 1 — 3D Array/Tensor
Tạo một mảng/tensor 3D có kích thước (2, 2, 2) chứa 8 giá trị. Truy cập từng phần tử riêng lẻ và thực hiện cắt lát.

```python
import numpy as np
import torch

# NumPy
arr_3d = np.arange(8).reshape(2, 2, 2)
print("3D Array:\n", arr_3d)
print("Element [0,1,1]:", arr_3d[0, 1, 1])
print("Slice [0, :, :]:\n", arr_3d[0, :, :])
print("Slice [:, 0, :]:\n", arr_3d[:, 0, :])

# PyTorch
tensor_3d = torch.arange(8).reshape(2, 2, 2)
print("\n3D Tensor:\n", tensor_3d)
print("Element [0,1,1]:", tensor_3d[0, 1, 1])
print("Slice [0, :, :]:\n", tensor_3d[0, :, :])
print("Slice [:, 0, :]:\n", tensor_3d[:, 0, :])
```

# Exercise 2 — ReLU Activation
Triển khai hàm kích hoạt ReLU (ReLU(x) = max(0, x)) sử dụng `np.maximum` cho NumPy và `torch.maximum` cho PyTorch.

```python
# NumPy
def relu_numpy(x):
    return np.maximum(0, x)

print("NumPy:", relu_numpy(np.array([1, -3, 2.5])))
print("NumPy single:", relu_numpy(-5))

# PyTorch
def relu_pytorch(x):
    return torch.maximum(torch.tensor(0.0), x)

print("PyTorch:", relu_pytorch(torch.tensor([1.0, -3.0, 2.5])))
print("PyTorch single:", relu_pytorch(torch.tensor(-5.0)))
```

# Exercise 3 — Euclidean Norm (1D Vector)
Viết một hàm tính chuẩn Euclid (L2 norm) của một vectơ 1D bằng cả NumPy (không dùng vòng lặp for) và PyTorch.

```python
# NumPy
def euclidean_norm_numpy(x):
    return np.sqrt(np.sum(x ** 2))

my_vector_np = np.array([0.5, -1.2, 3.3, 4.5])
print("NumPy:", euclidean_norm_numpy(my_vector_np))

# PyTorch
def euclidean_norm_pytorch(x):
    return torch.sqrt(torch.sum(x ** 2))

my_vector_pt = torch.tensor([0.5, -1.2, 3.3, 4.5])
print("PyTorch:", euclidean_norm_pytorch(my_vector_pt))
```

# Exercise 4 — Euclidean Norm (2D Matrix, Row-wise)
Viết một hàm tính chuẩn Euclid của từng hàng trong một ma trận 2D bằng cả NumPy và PyTorch.

```python
# NumPy
def euclidean_norm_2d_numpy(X):
    return np.sqrt(np.sum(X ** 2, axis=1))

my_matrix_np = np.array([[0.5, -1.2, 4.5],
                          [-3.2, 1.9, 2.7]])
print("NumPy:", euclidean_norm_2d_numpy(my_matrix_np))

# PyTorch
def euclidean_norm_2d_pytorch(X):
    return torch.sqrt(torch.sum(X ** 2, dim=1))

my_matrix_pt = torch.tensor([[0.5, -1.2, 4.5],
                              [-3.2, 1.9, 2.7]])
print("PyTorch:", euclidean_norm_2d_pytorch(my_matrix_pt))
```

# Exercise 5 — Iris Dataset Feature Means
Tính giá trị trung bình của các đặc trưng trong tập dữ liệu iris bằng cả NumPy và PyTorch.

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

# NumPy: compute mean of each feature (result should be an array of size 4)
feature_means_np = np.mean(X, axis=0)
print("NumPy feature means:", feature_means_np)

# PyTorch: convert X to a tensor and compute mean using dim=0
X_tensor = torch.tensor(X, dtype=torch.float64)
feature_means_pt = torch.mean(X_tensor, dim=0)
print("PyTorch feature means:", feature_means_pt)
```
